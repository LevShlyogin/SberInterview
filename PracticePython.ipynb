```python
import pandas as pd
import numpy as np
import ast

data = pd.read_excel('/content/Задание_2.xlsx', sheet_name='Исх.данные')
data.head()
```

```python
# Добавляем новый столбец
data['count'] =[2, 6, 2, 3, 2]

data.head()
```

```python
# Дублируем строки по значениям в новом столце
data_cop = data.loc[data.index.repeat(data['count'])].reset_index(drop=True)

data_cop
```

Функция для обработки столбцов

```python
def column_processing(col_name):
    # Получаем вложенный список
    nested_list = data_cop[col_name].iloc[0]
    # Приводим к типу list
    nested_list1 = ast.literal_eval(nested_list)

    # Разделяем на два отдельных списка
    list1 = nested_list1[0]
    list2 = nested_list1[1]

    # Заменяем значения в столбце
    data_cop.at[0, col_name] = list1
    data_cop.at[1, col_name] = list2

    #Получаем третий элемент таблицы
    nested_list2 = data_cop[col_name].iloc[4]

    #Приводим к типу list
    nested_list3 = ast.literal_eval(nested_list2)

    # Разделяем на пять отдельных списков
    list3 = nested_list3[0]
    list4 = nested_list3[1]
    list5 = nested_list3[2]
    list6 = nested_list3[3]
    list7 = nested_list3[4]
    list8 = nested_list3[5]

    # Заменяем значения в столбце
    data_cop.at[2, col_name] = list3
    data_cop.at[3, col_name] = list4
    data_cop.at[4, col_name] = list5
    data_cop.at[5, col_name] = list6
    data_cop.at[6, col_name] = list7
    data_cop.at[7, col_name] = list8

    # Получаем вложенный список
    nested_list4 = data_cop[col_name].iloc[8]

    #Приводим к типу list
    nested_list5 = ast.literal_eval(nested_list4)

    # Разделяем на два отдельных списка
    list9 = nested_list5[0]
    list10 = nested_list5[1]

    # Заменяем значения в столбце
    data_cop.at[8, col_name] = list9
    data_cop.at[9, col_name] = list10

    # Получаем вложенный список
    nested_list6 = data_cop[col_name].iloc[10]

    #Приводим к типу list
    nested_list7 = ast.literal_eval(nested_list6)

    # Разделяем на три отдельных списка
    list11 = nested_list7[0]
    list12 = nested_list7[1]
    list13 = nested_list7[2]

    # Заменяем значения в столбце
    data_cop.at[10, col_name] = list11
    data_cop.at[11, col_name] = list12
    data_cop.at[12, col_name] = list13

    # Получаем вложенный список
    nested_list8 = data_cop[col_name].iloc[13]

    #Приводим к типу list
    nested_list9 = ast.literal_eval(nested_list8)

    # Разделяем на два отдельных списка
    list14 = nested_list9[0]
    list15 = nested_list9[1]

    # Заменяем значения в столбце
    data_cop.at[13, col_name] = list9
    data_cop.at[14, col_name] = list10

    return data_cop
```

```python
# Функция для замены "true" на "True"
def fix_true(value):
    return value.replace("true", "True")

# Применяем функцию только к указанному столбцу
processed_data["owner"] = processed_data["owner"].apply(fix_true)

# Выводим результат
print(processed_data)
```
